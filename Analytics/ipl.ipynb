```python
deliveries = deliveries[deliveries['is_super_over'] == 0]
deliveries = deliveries[deliveries['dismissal_kind'] != 'run out']
```

```python
# deliveries = deliveries.loc[:, ['batsman', 'bowler', 'batsman_runs', 'total_runs', 'player_dismissed']]
```

```python
pairs = deliveries.loc[:, ['batsman', 'bowler']]
```

```python
dataset = pairs.drop_duplicates()
```

```python
# for batsman, bowler in dataset.values:
#     print(deliveries[deliveries['batsman'] == batsman]['batsman'] & deliveries['bowler'] == bowler])
```

```python
i = 0
for batsman, bowler in dataset.iloc[:, :2].values:
    i += 1
    print(i, batsman, bowler)
    dataset.loc[(dataset['batsman'] == batsman) & (dataset['bowler'] == bowler), ['batsman_runs']] = deliveries[(deliveries['batsman'] == batsman) & (deliveries['bowler'] == bowler)]['batsman_runs'].sum()
    dataset.loc[(dataset['batsman'] == batsman) & (dataset['bowler'] == bowler), ['total_runs']] = deliveries[(deliveries['batsman'] == batsman) & (deliveries['bowler'] == bowler)]['total_runs'].sum()
    dataset.loc[(dataset['batsman'] == batsman) & (dataset['bowler'] == bowler), ['dismissals']] = deliveries[(deliveries['batsman'] == batsman) & (deliveries['bowler'] == bowler)]['player_dismissed'].notnull().sum()
    dataset.loc[(dataset['batsman'] == batsman) & (dataset['bowler'] == bowler), ['balls']] = len(pairs[(pairs['batsman'] == batsman) & (pairs['bowler'] == bowler)])
```

```python
dataset.drop(['strike_rate', 'economy'], axis=1, inplace=True)
```

```python
dataset['batsman_runs'] = np.nan
dataset['total_runs'] = np.nan
dataset['dismissals'] = np.nan
dataset['balls'] = np.nan
# dataset['strike_rate'] = np.nan
# dataset['economy'] = np.nan
```

```python
dataset['strike_rate'] = dataset['batsman_runs'] * 100 / dataset['balls']
dataset['economy'] = dataset['total_runs'] * 6 / dataset['balls']
dataset['bowler_strike_rate'] = dataset['balls'] / dataset['dismissals'].replace(0,0.1)
dataset['bowler_average'] = dataset['total_runs'] / dataset['dismissals'].replace(0,0.1)
```
